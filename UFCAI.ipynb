```python
import pandas as pd
import numpy as np

fight_stats = pd.read_csv("ufc_fight_stats.csv")
fight_results = pd.read_csv("ufc_fight_results.csv")
fighter_profile = pd.read_csv("ufc_fighter_profile.csv")
```

```python
fight_stats.head()

```

```python
fight_results['EVENT'] = fight_results['EVENT'].apply(lambda x: x.replace(' ', ''))
fight_results['BOUT'] = fight_results['BOUT'].apply(lambda x: x.replace(' ', ''))

fight_stats['EVENT'] = fight_stats['EVENT'].apply(lambda x: x.replace(' ', ''))
fight_stats['BOUT'] = fight_stats['BOUT'].apply(lambda x: x.replace(' ', ''))


fight_results.head()
# UFC Fight Night: Vettori vs. Dolidze 2
# UFC Fight Night: Vettori vs. Dolidze 2
```

```python
fight_total = fight_stats.merge(fight_results, on=["EVENT", "BOUT"], how="inner").merge(fighter_profile, on="FIGHTER", how="inner")
fight_total.drop(columns=['URL_y', 'URL_x'], inplace=True)
fight_total.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
feature_names = ['EVENT', 'BOUT', 'STAT_ROUND', 'FIGHTER', 'KD', 'SIG.STR.', 'SIG.STR. %', 'TOTAL STR.',
                  'TD', 'TD %', 'SUB.ATT', 'REV.', 'CTRL', 'HEAD', 'BODY', 'LEG', 'DISTANCE', 'CLINCH', 'GROUND']
```

```python
# One-Hot Encode categorical variables
fight_total_encoded = pd.get_dummies(fight_total, columns=['TIME FORMAT', 'STANCE', 'METHOD', ])

print(fight_total_encoded.info())
```

```python

# Define features and target
X = fight_total_encoded[feature_names]
y = fight_total_encoded["WINNER"]

# Split the data into training and testing sets
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.2, random_state=1)

# Train the Decision Tree Regressor model
dtr_model = DecisionTreeRegressor(random_state=1)
dtr_model.fit(train_X, train_y)

# Make predictions
predictions = dtr_model.predict(test_X)

mse = mean_squared_error(test_y, predictions)
r2 = r2_score(test_y, predictions)

print("Mean Squared Error: ", mse)
print("R-squared: ", r2)
```
